Normalize data with MinMaxScaler

This technique re-scales the data between a specified range(in this case, between 01), to ensure that certain features do not affect the final prediction more than the other features.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
X_trans = scaler.fit_transform(X)
```

```python
X_trans
```

```python
y=pd.Series(y)
y
```

```python
# enumerate splits     10 Splits
cv = KFold(n_splits=10, random_state=4, shuffle=False)
for trained, tested in cv.split(X_trans):
    print("Train Index: ", trained, "\n")
    print("Test Index: ", tested)
```

```python
X_train, X_test, y_train, y_test = X_trans[trained], X_trans[tested], y[trained], y[tested]
```
